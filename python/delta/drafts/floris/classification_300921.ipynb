### Petit point sur les "modes" (variable `input_structure`) :

> On teste deux structurations pour les données d'entrée du modèle dans ce script :
> 
> - **Mode "2D" :**
> 
>     Les données fournies au modèle sont en deux dimensions uniquement : on "empile" les courbes Réf, G, A, M, k, l dans une seule dimension, ce qui donne une numpy array de dimensions : (N,304,6) où N l'effectif (taille du batch), avec 304 points pour 6 courbes.
> 
>     Pour que les données soient utilisables par tensorflow, on est obligé de rajouter une nouvelle dimension, les données sont donc de dimension finale : (N,304,1,6).
> 
>
> - **#### Mode "3D" :**
> 
>     Les données fournies au modèle sont en 3 dimensions : on va "empiler" les courbes G, A, M, k, l (pas la courbe de référence) dans le premier axe ; puis on "empile" la courbe de référence au-dessus de chacune de ces courbes, dans une deuxième dimension (un deuxième axe) (cette courbe de référence est donc répétée 5 fois) : soit une dimension finale pour la numpy array : (N,304,5,2).
> 
>     L'idée dans cette structuration est que chaque courbe G, A, M, k, l est "lue" par le modèle indépendamment des autres, mais chacune en rapport avec la courbe de référence (qui est "empilée" dans un autre axe). Pour le modèle qui analyse ces données
> 
>     On va initialement utiliser des couches de convolution avec un noyau (_kernel_) de taille (4,1) : c'est-à-dire que l'opération de convolution est réalisée parallèlement pour chaque courbe (G,A,M,k,l) mais prend en compte à chaque fois cette courbe & la courbe de référence (puisque ces deux courbes sont empilées sur l'axe des channels).
> 
>     Dans les derniers blocs de convolution, on utilise des noyaux de taille (4,5) : ainsi, l'opération de convolution va s'appliquer une seule fois pour l'intégralité des courbes, mais prendre en compte simultanément les motifs détectés sur les 5 courbes G, A, M, k, l.
> 
>     Empiriquement, cette solution semble donner les meilleurs résultats : c'est la solution par défaut qui est choisie dans ce script.

On va ensuite créer les différentes classes nécessaires au formattage des données pour l'entraînement et à l'augmentation.

Le générateur `ITGenerator` pourra être instancié avec les données x et y du jeu souhaité (train, val, test), et va permettre de renvoyer des batches en formattant les données selon l'architecture choisie (2D ou 3D). De plus, le générateur accepte aussi comme argument un algorithme d'augmentation permettant d'augmenter les données des courbes.

Pour l'augmentation, on crée une classe abstraite générique `ITAugmenter` qui déclare une fonction `augment`, puis un crée une classe fille `ITAlgorithmicAugmenter` qui permet une augmentation mathématique simple des courbes (décalage +- bruit).

Si nécessaire, on pourra ainsi implémenter une classe d'augmentation plus aboutie, par exemple basée sur l'utilisation d'un GAN.

```python
# on crée une classe abstraite pour l'augmentation des données
class ITAugmenter():
    def __init__(self):
        pass
    
    def augment(self, batch_x, batch_y):
        assert False, "This method (ITAugmenter.augment) should be implemented in child class"
        # return batch_x.copy(), batch_y.copy()
        return batch_x, batch_y

# on crée une classe d'augmentation fille qui va hériter de l'augmenteur générique
# cette structure pourra être utile si besoin d'ajouter d'autres types d'augmentation
class ITAlgorithmicAugmenter(ITAugmenter):
    def __init__(self, add_noise=None, add_shifting=None, seed=42):
        super().__init__()
        self.add_noise          = add_noise
        self.add_shifting       = add_shifting
        self.rng                = np.random.RandomState(seed)
        
    def augment(self, batch_x, batch_y):
        # batch_x, batch_y = batch_x.copy(), batch_y.copy()
        if self.add_noise is not None:
            freq=self.add_noise['freq'] # between 0. and 1., how often do wee add noise
            for i in range(batch_x.shape[0]):
                if self.rng.random(1) < freq:
                    std=self.rng.uniform(low=self.add_noise['minstd'], high=self.add_noise['maxstd'], size=1)
                    batch_x[i,...] += self.rng.normal(loc=0, scale=std, size=batch_x.shape[1:])
        if self.add_shifting is not None:
            freq=self.add_shifting['freq'] # between 0. and 1., how often do wee add noise
            for i in range(batch_x.shape[0]):
                if self.rng.random(1) < freq:
                    d=self.rng.choice(np.arange(self.add_shifting['min'],self.add_shifting['max']))
                    if self.rng.random() > .5: # left
                        batch_x[i,...] = np.concatenate([batch_x[i,d:,...],np.zeros_like(batch_x[i,:d,...])], axis=0)
                    else: # right
                        batch_x[i,...] = np.concatenate([np.zeros_like(batch_x[i,-d:,...]),batch_x[i,:-d,...]], axis=0)
        return batch_x, batch_y
        
# puis on crée la classe générateur proprement dit
# cette classe hérite du générateur générique de keras (tensorflow)
class ITGenerator(tf.keras.utils.Sequence):
    def __init__(self, x, y, batch_size = 8, shuffle=True, augmenter=None, seed=42, output="3D", allow_incomplete_batches = False):
        self.x                  = x             # the x array
        self.y                  = y             # the y array
        self.batch_size         = batch_size    # the batch size
        self.shuffle            = shuffle       # set to true for shuffling samples
        self.augmenter          = augmenter
        self.seed               = seed
        self.output             = output
        self.rng                = np.random.RandomState(self.seed)
        self.allow_incomplete_batches = allow_incomplete_batches
        self.on_epoch_end()

    # cette fonction doit renvoyer le nombre de batches réalisables par epoch avec les données du générateur
    def __len__(self): # should return the number of batches per epoch
        if self.allow_incomplete_batches:
            return int(np.ceil(self.x.shape[0]/self.batch_size))
        return self.x.shape[0]//self.batch_size

    # cette fonction est appelée par model.fit() à la fin de chaque epoch
    # l'objectif est de 'réinitialiser' le générateur afin qu'il re-mélange les échantillons
    # pour les présenter dans un ordre différent et avec une augmentation différente
    def on_epoch_end(self): # shuffle samples
        self.indexes = np.arange(self.x.shape[0])
        if self.shuffle:
            self.rng.shuffle(self.indexes)

    # cette fonction permet de renvoyer un batch (x et y) pour l'entraînement ou l'inférence
    def __getitem__(self, index):
        # get indexes for current batch
        b_start = index*self.batch_size
        b_end = np.minimum(((index+1)*self.batch_size), len(self.indexes))
        batch_indexes = self.indexes[b_start:b_end]
        # get data
        batch_x = self.x[batch_indexes,...].copy()
        batch_y = self.y[batch_indexes,...].copy()
        # augment id needed
        if self.augmenter is not None:
            batch_x, batch_y = self.augmenter.augment(batch_x, batch_y)
        
        if self.output == "2D":
            # reshape in 2D (add 1 dummy dimension)
            batch_x = np.expand_dims(batch_x, axis=-2)
        elif self.output == "3D":
            # reshape in 3D
            # take reference curves, repeat them and stack them with GAMkl curves
            batch_x = np.concatenate([np.expand_dims(np.tile(batch_x[...,[0]],5),-1), np.expand_dims(batch_x[...,1:],-1)], -1)
        else:
            assert False, "Unknown output structure: {}".format(self.output)
        
        return batch_x, batch_y
    
# create our augmenter
standard_augmenter = ITAlgorithmicAugmenter(# add_noise=None,
                                            add_noise=dict(freq=.2, minstd=.001, maxstd=.002),
                                            # add_shifting=None,
                                            add_shifting=dict(freq=.8, min=1, max=4),
                                            seed = 0)

# create our train generator
train_gen = ITGenerator(x=x_train, y=y_train, batch_size=batch_size, augmenter = standard_augmenter, output = input_structure, seed = 0)
valid_gen = ITGenerator(x=x_valid, y=y_valid, batch_size=batch_size, augmenter = None, output = input_structure, seed = 0, shuffle=False, allow_incomplete_batches=True)
test_gen = ITGenerator(x=x_test, y=y_test, batch_size=batch_size, augmenter = None, output = input_structure, seed = 0, shuffle=False, allow_incomplete_batches=True)
```

## 2. Création du modèle et entraînement

### 2.1. Création du modèle

En premier pour cette partie, nous allons créer la structure du modèle. D'abord, on crée des fonctions qui vont permettre d'instancier le modèle, en fonction des différents hyperparamètres choisis en début de script.

Notre modèle sera un simple CNN basé sur la répétition de blocs de convolution (Conv2D -> BatchNorm -> ReLU x 2) avec des couches de MaxPooling2D entre chaque bloc.

Le nombre de filtres des couches de convolution est multiplié par deux à chaque bloc : 32, 64, 128, 256, 512.

On termine le modèle par une couche de GlobalMaxPooling2D puis une couche Dense avec 8 neurones pour les 8 classes à prédire.

```python
# on commence par réaliser une sous-routine qui permet d'ajouter un "bloc" de convolution,
# c'est à dire un enchaînement Conv2D -> Batch Norm -> ReLU, répété deux fois
def conv1d_block(input_tensor, n_filters, kernel_size_width=3, kernel_size_height=1, batchnorm=True):
    # first layer
    #x = Conv1D(filters=n_filters, kernel_size=kernel_size, kernel_initializer="he_normal", padding="same", data_format='channels_last')(input_tensor)
    x = tf.keras.layers.Conv2D(filters=n_filters, kernel_size=(kernel_size_width,kernel_size_height), kernel_initializer="he_normal", padding="same", data_format='channels_last')(input_tensor)
    if batchnorm:
        x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("relu")(x)
    # second layer
    #x = Conv1D(filters=n_filters, kernel_size=kernel_size, kernel_initializer="he_normal", padding="same", data_format='channels_last')(x)
    x = tf.keras.layers.Conv2D(filters=n_filters, kernel_size=(kernel_size_width,kernel_size_height), kernel_initializer="he_normal", padding="same", data_format='channels_last')(x)
    if batchnorm:
        x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("relu")(x)
    return x

# puis on réalise la routine qui crée le modèle entier
def get_classnet_custom(input_signal, blocks=5, wide_kernel_starts_at_block=-1, n_filters=32, kernel_size=3, dropout=0.1, batchnorm=True, n_classes=8):
    x = input_signal

    # pour chaque bloc de convolution :
    # on ajoute un bloc (conv>batchnorm>relu>conv>batchnorm>relu) avec n filtres
    # puis on poursuit par une couche de maxpooling pour diviser la résolution des feature maps par 2
    # puis on poursuit avec le prochain bloc, avec n*2 filtres à convolution
    # et ainsi de suite jusqu'au nombre désiré de blocs (4~5 le plus souvent)
    #
    # définition de 'wide_kernel_starts_at_block':
    # dans notre structure de données, les courbes (axe y) sont "alignées" en parallèle sur l'axe y
    # si l'on choisit un kernel de taille par ex. : (4,1),
    # alors l'opération de convolution va se faire en parallèle pour chaque courbe
    # si l'on choisit un kernel de taille (4,5)
    # alors l'opération de convolution va prendre en compte les 5 courbes (G, A, M, k, l) en parallèle
    # ici, on choisit de commencer par des blocs de convolution (4,1), c'est-à-dire que les différentes courbes
    # sont traitées chacune de leur côté
    # l'objectif est de repérer des motifs de pics indépendamment sur chaque courbe
    # dans les derniers blocs à l'inverse, on va réaliser des convolutions (4,5)
    # afin de prendre en compte simultanément la présence de motifs pathologiques sur les différentes courbes
    for b in range(blocks):
        if wide_kernel_starts_at_block==-1:
            kernel_size_height = 1
        elif b>=wide_kernel_starts_at_block-1:
            kernel_size_height = 5
        else:
            kernel_size_height = 1
        x = conv1d_block(x, n_filters=n_filters*np.power(2,b), kernel_size_width=kernel_size, kernel_size_height=kernel_size_height, batchnorm=batchnorm)
        if b<blocks-1:
            x = tf.keras.layers.MaxPooling2D((2,1)) (x)

    # on termine par une couche de global max pooling plutôt que d'utiliser une couche flatten (cf par ex Inception v3)
    x = tf.keras.layers.GlobalMaxPooling2D() (x)
    # puis une couche de Dense avec 8 neurones pour les 8 classes à prédire
    outputs = tf.keras.layers.Dense(n_classes, activation='softmax') (x)

    # on peut alors instancier le modèle
    model = tf.keras.models.Model(inputs=[input_signal], outputs=[outputs])

    return model
```

Puis on peut instancier le modèle proprement dit en lui fournissant les hyperparamètres choisis :

```python
# on crée le placeholder pour l'input
if input_structure == "2D":
    input_signal = tf.keras.layers.Input((spe_width, 1, 6), name='input_if')
    wide_kernel_starts_at_block = 999 # pas besoin si 2D
elif input_structure == "3D":
    input_signal = tf.keras.layers.Input((spe_width, 5, 2), name='input_if')
    
# on construit le modèle depuis l'input
model = get_classnet_custom(input_signal,
                            blocks=blocks,
                            wide_kernel_starts_at_block=wide_kernel_starts_at_block,
                            n_filters=filters,
                            kernel_size=kernel_size,
                            dropout=dropout,
                            batchnorm=batchnorm,
                            n_classes=len(y_labels))

# on crée des callbacks pour monitorer l'entraînement :
# - model checkpoint -> permet de sauvegarder les poids du modèle quand la val loss s'améliore
# - reduce LR on plateau -> permet de diminuer le learning rate quand la val loss stagne (3 epochs sans amélioration)
# - early stopping -> permet d'arrêter l'entraînement quand la val loss stagne (10 epochs sans amélioration)
callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, min_delta=1e-3, verbose=verbose, restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3, min_delta=1e-3, min_lr=min_lr, verbose=verbose),
    tf.keras.callbacks.ModelCheckpoint(os.path.join(path_out,model_name), monitor='val_loss', verbose=verbose, save_best_only=True, save_weights_only=False)
]

# on compile le modèle (i.e. on ajoute la loss, les métriques et on instancie l'algorithme d'optimisation)
model.compile(loss=tf.keras.losses.CategoricalCrossentropy(), optimizer=tf.keras.optimizers.Adam(lr=base_lr), metrics=[tf.keras.metrics.CategoricalAccuracy()])
```

On peut alors vérifier l'intégrité de la structure de notre modèle :

```python
print(model.summary())
```

On va alors pouvoir lancer l'entraînement :

```python
N_EPOCHS = 1000 # on fixe volontairement le nombre max d'epochs très élevé car il y a un callback d'early stopping

print('Setting batch size to: '+str(batch_size))
print('Setting maximal number of epochs to: '+str(N_EPOCHS))

# on lance l'entraînement proprement dit
results = model.fit(train_gen,
                    batch_size=batch_size,
                    epochs=N_EPOCHS,
                    callbacks=callbacks,
                    verbose=verbose,
                    validation_data=valid_gen,
                    validation_steps = valid_gen.__len__())

# à la fin de l'entraînement, on sauvegarde les logs (monitoring des losses/métriques)
with open(os.path.join(path_out,log_name), 'wb') as file_pi:
    pickle.dump(results.history, file_pi)
```

## 3. Etapes post-entraînement : visualisation et validation

Les prédictions du modèle sont "one-hot encoded". Par exemple pour un batch de 4 échantillons et avec 8 classes, la prédiction du modèle pourrait être :

| Echantillon | Classe 1 | Classe 2 | Classe 3 | Classe 4 | Classe 5 | Classe 6 | Classe 7 | Classe 8 |

| 1 | 0 | 0 | 0.95 | 0 | 0 | 0.05 | 0 | 0 |
| 2 | 0.99 | 0 | 0 | 0.01 | 0 | 0 | 0 | 0 |
| 3 | 0 | 0.32 | 0 | 0.28 | 0 | 0.24 | 0.16 | 0 |
| 4 | 0 | 0 | 0 | 0 | 0 | 1.00 | 0 | 0 |

Dans cet exemple, les 4 échantillons peuvent dont être classés respectivement en classe : 3, 1, 2, 6

On va donc ci-dessous convertir les prédictions "brutes" en classes textuelles selon cette méthodologie :

```python
pred_labels = [y_labels[v][12:] for v in np.argmax(test_preds_, axis=1)]
real_labels = [y_labels[v][12:] for v in np.argmax(y_test, axis=1)]
```

On peut afficher un extrait de la matrice one hot encoded (= prédictions brutes) et des labels textuelles correspondantes :

```python
print("Prédictions brutes:")
print(np.round(test_preds_[:4,...],3))
print("")

print("Ordre des classes :")
print([label[12:] for label in y_labels])
print("")

print("Prédictions textuelles:")
print(pred_labels[:4])
```

Enfin on peut calculer et afficher le tableau de contingence, la précision et la sensibilité/spécificité :

```python
# on détermine la matrice de contingence :
contingency_table = pd.crosstab(pd.Series(real_labels), pd.Series(pred_labels))
contingency_table.index.name = "Ground truth:"
contingency_table.columns.name = "Predicted:"

# on calcule la précision par classe "ground truth" :
pr_by_groundtruth_class = np.array([contingency_table.loc[c,c] for c in contingency_table.columns]) / np.sum(np.array(contingency_table), axis=1)
# et on ajoute ça en colonne au tableau de contingence :
contingency_table.loc[:,"Accuracy (by ground truth class)"] = ["{:.1f} %".format(100*v) for v in pr_by_groundtruth_class]

pr = np.sum(np.array(pred_labels)==np.array(real_labels))/len(pred_labels) # on calcule la précision globale

# on calcule les vrais & faux positifs et négatifs (tp, tn, fp, fn) en fonction de "normal" ou pas "normal"
tp = np.sum( (np.array(pred_labels)!="Normal") & (np.array(real_labels)!="Normal") )
fp = np.sum( (np.array(pred_labels)!="Normal") & (np.array(real_labels)=="Normal") )
tn = np.sum( (np.array(pred_labels)=="Normal") & (np.array(real_labels)=="Normal") )
fn = np.sum( (np.array(pred_labels)=="Normal") & (np.array(real_labels)!="Normal") )
# on peut alors calculer sensibilité & spécificité pour la question "Normal vs. non Normal"
se = tp/(tp+fn)
sp = tn/(tn+fp)

display(contingency_table)

print("Précision globale : {:.1f} %".format(100*pr))
print("")
print("Normal vs. anormal :")
print("    Sensibilité : {:.1f} %".format(100*se))
print("    Spécificité : {:.1f} %".format(100*sp))
```
