### 2.2. Extraction des données X et Y

Après avoir chargé le fichier .csv, on va en extraire les informations qui nous intéressent.

On commencer par lister les échantillons (i.e. groupes de 6 courbes + 1 annotation textuelle) contenus dans le dataset :

```python
# le pattern qui va permettre de déterminer à quel échantillon & quelle piste correspond chaque colonne :
column_regex_pattern = "^([0-9a-zA-Z]+)-(ELP|G|A|M|K|L)(-y)?$"

# pour chaque colonne, on extrait uniquement l'ID échantillon, afin de lister les échantillons du jeu de données :
sample_columns = raw_dataset.columns[1:].tolist() # la liste des noms de colonnes du dataset
# on utilise la règle regex pour convertir NOM DE COLONNE -> ID ECHANTILLON
samples_list = np.unique([re.sub(column_regex_pattern, "\\1", column) for column in sample_columns]).tolist()
# dans samples_list on a donc la liste des échantillons retrouvés

print("Found {} samples in dataset.".format(len(samples_list)))
```

Ensuite, pour chaque échantillon, on va récupérer ses informations et les stocker dans une liste.

On récupère par la même occasion les données des maps de segmentation.

```python
samples_data_X = [] # cette liste contiendra pour chaque échantillon une numpy array de dimension (300,6) contenant
# les valeurs des 6 courbes de 300 points
samples_data_Y = [] # cette liste contiendra pour chaque échantillon la label textuelle (= une str)
samples_data_Ymap = [] # cette liste contiendra pour chaque échantillon les labels maps

for sample_ID in samples_list: # pour chaque échantillon
    
    # on va alors pouvoir récupérer les données d'entrée et de sortie pour chaque échantillon
    tracks_data = [] # la liste des données de chaque piste
    tracks_data_y = []
    
    # on récupère les données X :
    for track in INPUT_TRACKS_ORDER: # pour chaque piste (ELP, G, etc.)
        tracks_data.append(raw_dataset.loc[:,sample_ID+"-"+track].iloc[1:].to_numpy()) # on récupère la liste des valeurs
        # de la courbe puis on convertir ça en numpy array et on la stocke dans le bon ordre
        
    # on récupère les données Y(maps):
    for track in OUTPUT_TRACKS_ORDER: # pour chaque piste (G, A, etc.)
        tracks_data_y.append(raw_dataset.loc[:,sample_ID+"-"+track+"-y"].iloc[1:].to_numpy()) # on récupère la liste des valeurs
        # de la courbe puis on convertir ça en numpy array et on la stocke dans le bon ordre

    # ensuite on va "stacker" ces courbes entre elles pour obtenir une numpy array de plus haute dimension
    tracks_data = np.stack(tracks_data, axis=-1) # on passe d'une liste de 6 arrays de dimension (300,) à une array de dimension (300,6)
    tracks_data_y = np.stack(tracks_data_y, axis=-1) # on passe d'une liste de 6 arrays de dimension (300,) à une array de dimension (300,6)
    
    # et on récupère l'annotation (label) :
    label_data = raw_dataset.loc[:,sample_ID+"-ELP"].iloc[0]
    
    # et enfin on stocke tout ça dans notre liste initiale :
    samples_data_X.append(tracks_data)
    samples_data_Y.append(label_data)
    samples_data_Ymap.append(tracks_data_y)
```

```python
print("La liste 'samples_data_X' contient {} numpy arrays de dimension {}".format(len(samples_data_X), samples_data_X[0].shape))
print("Exemple de X pour le premier échantillon:")
print(samples_data_X[0])

print("")

print("La liste 'samples_data_Y' contient {} chaînes de caractères".format(len(samples_data_Y)))
print("Exemple de Y pour le premier échantillon:")
print(samples_data_Y[0])

print("")

print("La liste 'samples_data_Ymap' contient {} numpy arrays de dimension {}".format(len(samples_data_Ymap), samples_data_Ymap[0].shape))
print("Exemple de Y pour le premier échantillon:")
print(samples_data_Ymap[0])
```

Enfin, on va stacker les numpy arrays du X de tous les échantillons de dimensions `(300,6,)` en une seule numpy array de dimension `(N,300,6)` où `N` = le nombre d'échantillons.

On va également convertir les valeurs en numérique (`float32`).

On réalise la même chose pour les annotations Y (maps) pour obtenir une numpy array de dimension `(N,300,5)` (il n'y a pas d'annotation pour la courbe de référence).

```python
X_dataset = np.stack(samples_data_X, axis=0)

X_dataset = X_dataset.astype(np.float32)

print("Final X dataset shape: {}".format(X_dataset.shape))

Ymap_dataset = np.stack(samples_data_Ymap, axis=0)

Ymap_dataset = Ymap_dataset.astype(np.float32)

print("Final Y(maps) dataset shape: {}".format(Ymap_dataset.shape))
```

### 2.5. Générateurs, augmentation et conversion des données pour import dans DELTA

Cette partie est spécifique de DELTA : on va convertir les données X et Y afin qu'elles soient interprétables par le modèle DELTA.

Cela est valable autant pour l'entraînement que pour l'inférence.

On va pour cela utiliser un générateur qui va se charger :
1. De mélanger les échantillons aléatoirement à chaque epoch
2. De normaliser les courbes entre 0 et 1
3. De réaliser la data augmentation
4. De s'assurer que la dimension des numpy arrays fournies au modèle correspond à la dimension attendue par celui-ci

> ### Petit point sur la dimension des données d'entrée pour le modèle de deep learning :  
> 
> Les données fournies au modèle sont en 3 dimensions : on va "empiler" les courbes G, A, M, K, L (pas la courbe de référence) dans le premier axe ; puis on "empile" la courbe de référence au-dessus de chacune de ces courbes, dans une deuxième dimension (un deuxième axe) (cette courbe de référence est donc répétée 5 fois) : soit une dimension finale pour la numpy array : `(N,304,5,2)` avec N le nombre d'échantillons.
> 
> L'idée dans cette structuration est que chaque courbe G, A, M, K, L est "lue" par le modèle indépendamment des autres, mais chacune en rapport avec la courbe de référence (qui est "empilée" dans un autre axe).
> 
> On va initialement utiliser des couches de convolution avec un noyau (_kernel_) de taille `(4,1)` : c'est-à-dire que l'opération de convolution est réalisée parallèlement pour chaque courbe (G, A, M, K, L) mais prend en compte à chaque fois cette courbe & la courbe de référence (puisque ces deux courbes sont empilées sur l'axe des channels).
> 
> Dans les derniers blocs de convolution, on utilise des noyaux de taille `(4,5)` : ainsi, l'opération de convolution va s'appliquer une seule fois pour l'intégralité des courbes, mais prendre en compte simultanément les motifs détectés sur les 5 courbes G, A, M, K, L.

On va donc créer les différentes classes nécessaires au formattage des données pour l'entraînement et à l'augmentation.

Le générateur `ITGenerator` pourra être instancié avec les données x et y du jeu souhaité (train, val, test), et va permettre de renvoyer des batches en formattant les données selon l'architecture choisie (2D ou 3D). De plus, le générateur accepte aussi comme argument un algorithme d'augmentation permettant d'augmenter les données des courbes.

Pour l'augmentation, on crée une classe abstraite générique `ITAugmenter` qui déclare une fonction `augment`, puis un crée une classe fille `ITAlgorithmicAugmenter` qui permet une augmentation mathématique simple des courbes (décalage &plusmn; bruit).

Si nécessaire, on pourra ainsi implémenter une classe d'augmentation plus aboutie, par exemple basée sur l'utilisation d'un GAN.

Il est à noter que le décalage des courbes n'apporte aucun avantage pour un FCN (fully convolutional network).

> **Important !**
>
> Par rapport au modèle de classification, on inclut également ici les maps y.
> La différence la plus notable est qu'ici l'algorithme d'augmentation permet également de pouvoir rajouter aléatoirement des pics sur les échantillons (rajout d'un pic sur le X et de l'annotation correspondante sur le y). Néanmoins, on constate empiriquement que cette option n'apporte pas un gain significatif de performance, on ne s'en servira donc pas ici.

```python
# on crée une classe abstraite pour l'augmentation des données
class ITAugmenter():
    def __init__(self):
        pass
    
    def augment(self, batch_x, batch_y):
        assert False, "This method (ITAugmenter.augment) should be implemented in child class"
        # return batch_x.copy(), batch_y.copy()
        return batch_x, batch_y

class ITAlgorithmicAugmenter(ITAugmenter):
    def __init__(self, add_fake_mspikes=None, add_fake_flcs=None, add_noise=None, add_shifting=None, do_not_cover=True, seed=42):
        super().__init__()
        self.add_fake_mspikes   = add_fake_mspikes
        self.add_fake_flcs      = add_fake_flcs
        self.add_noise          = add_noise
        self.add_shifting       = add_shifting
        self.do_not_cover       = do_not_cover # if set to true, won't add a m-spike at a location where a m-spike already exists
        self.rng                = np.random.RandomState(seed)

    def augment(self, batch_x, batch_y):
        # batch_x, batch_y = batch_x.copy(), batch_y.copy()
        if self.add_noise is not None:
            freq=self.add_noise['freq'] # between 0. and 1., how often do wee add noise
            for i in range(batch_x.shape[0]):
                if self.rng.random(1) < freq:
                    std=self.rng.uniform(low=self.add_noise['minstd'], high=self.add_noise['maxstd'], size=1)
                    batch_x[i,...] += self.rng.normal(loc=0, scale=std, size=batch_x.shape[1:])
        if self.add_shifting is not None:
            freq=self.add_shifting['freq'] # between 0. and 1., how often do wee add noise
            for i in range(batch_x.shape[0]):
                if self.rng.random(1) < freq:
                    d=self.rng.choice(np.arange(self.add_shifting['min'],self.add_shifting['max']))
                    if self.rng.random() > .5: # left
                        batch_x[i,...] = np.concatenate([batch_x[i,d:,...],np.zeros_like(batch_x[i,:d,...])], axis=0)
                        batch_y[i,...] = np.concatenate([batch_y[i,d:,...],np.zeros_like(batch_y[i,:d,...])], axis=0)
                    else: # right
                        batch_x[i,...] = np.concatenate([np.zeros_like(batch_x[i,-d:,...]),batch_x[i,:-d,...]], axis=0)
                        batch_y[i,...] = np.concatenate([np.zeros_like(batch_y[i,-d:,...]),batch_y[i,:-d,...]], axis=0)
        if self.add_fake_mspikes is not None:
            freq=self.add_fake_mspikes['freq'] # between 0. and 1., how often do wee add noise
            for i in range(batch_x.shape[0]):
                if self.rng.random(1) < freq:
                    # add a fake mspike
                    for ms in range(self.rng.choice(np.arange(self.add_fake_mspikes['mincount'],self.add_fake_mspikes['maxcount']+1))):
                        # choose random dims
                        heavy_dim = self.rng.choice([1,2,3])
                        light_dim = self.rng.choice([4,5])
                        pos=self.rng.choice(np.arange(self.add_fake_mspikes['minpos'],self.add_fake_mspikes['maxpos']))
                        height=self.rng.uniform(self.add_fake_mspikes['minheight'],self.add_fake_mspikes['maxheight'])
                        width=self.rng.uniform(self.add_fake_mspikes['minwidth'],self.add_fake_mspikes['maxwidth'])
                        curve_with_mspike = norm.pdf(np.arange(304),pos,width)*height
                        curve_with_mspike_y = ((curve_with_mspike)>(np.max(curve_with_mspike)/10))*1
                        if self.do_not_cover:
                            # check if covering an already existing m-spike
                            if np.any(np.logical_and(curve_with_mspike_y==1, batch_y[i,...].max(1)==1)):
                                continue
                        for dim in range(batch_x.shape[-1]):
                            if (dim==heavy_dim) | (dim==light_dim): # do not add! but add to y
                                batch_y[i,...,dim-1] = np.maximum(batch_y[i,...,dim-1], curve_with_mspike_y)
                            else:
                                batch_x[i,...,dim] += curve_with_mspike + curve_with_mspike*np.random.normal(0,.01,curve_with_mspike.shape)
        if self.add_fake_flcs is not None:
            freq=self.add_fake_flcs['freq'] # between 0. and 1., how often do wee add noise
            for i in range(batch_x.shape[0]):
                if self.rng.random(1) < freq:
                    # add a fake mspike
                    for ms in range(self.rng.choice(np.arange(self.add_fake_flcs['mincount'],self.add_fake_flcs['maxcount']+1))):
                        # choose random dims
                        light_dim = self.rng.choice([4,5])
                        pos=self.rng.choice(np.arange(self.add_fake_flcs['minpos'],self.add_fake_flcs['maxpos']))
                        height=self.rng.uniform(self.add_fake_flcs['minheight'],self.add_fake_flcs['maxheight'])
                        width=self.rng.uniform(self.add_fake_flcs['minwidth'],self.add_fake_flcs['maxwidth'])
                        curve_with_mspike = norm.pdf(np.arange(304),pos,width)*height
                        curve_with_mspike_y = ((curve_with_mspike)>(np.max(curve_with_mspike)/10))*1
                        if self.do_not_cover:
                            # check if covering an already existing m-spike
                            if np.any(np.logical_and(curve_with_mspike_y==1, batch_y[i,...].max(1)==1)):
                                continue
                        for dim in range(batch_x.shape[-1]):
                            if dim==light_dim: # do not add! but add to y
                                batch_y[i,...,dim-1] = np.maximum(batch_y[i,...,dim-1], curve_with_mspike_y)
                            else:
                                batch_x[i,...,dim] += curve_with_mspike + curve_with_mspike*np.random.normal(0,.01,curve_with_mspike.shape)
        return batch_x, batch_y

# puis on crée la classe générateur proprement dit
# cette classe hérite du générateur générique de keras (tensorflow)
class ITGenerator(tf.keras.utils.Sequence):
    def __init__(self, x, y, batch_size = 8, shuffle=True, augmenter=None, seed=42, allow_incomplete_batches = False):
        self.x                  = x             # the x array
        self.y                  = y             # the y array
        self.batch_size         = batch_size    # the batch size
        self.shuffle            = shuffle       # set to true for shuffling samples
        self.augmenter          = augmenter     # la classe d'augmentation des données
        self.seed               = seed          # la seed
        self.rng                = np.random.RandomState(self.seed)
        self.allow_incomplete_batches = allow_incomplete_batches # cf. fonction __len__
        self.on_epoch_end()
        # add padding to curves (300 pts -> 304 points)
        self.add_padding_to_curves()
        
    def add_padding_to_curves(self):
        # on ajoute des 0 avant/après les courbes pour respecter l'entrée sur 304 points du modèle
        padding_total = SPE_WIDTH - self.x.shape[1]
        if padding_total<0:
            assert False, "Curves must be at least {} points wide".format(SPE_WIDTH)
        if padding_total>0:
            padding_before = padding_total//2
            padding_after = padding_total-padding_before
            self.x = np.concatenate( [ np.zeros((self.x.shape[0], padding_before, self.x.shape[-1])) ,
                                       self.x ,
                                       np.zeros((self.x.shape[0], padding_after,  self.x.shape[-1])) , ], axis=1 )
            self.y = np.concatenate( [ np.zeros((self.y.shape[0], padding_before, self.y.shape[-1])) ,
                                       self.y ,
                                       np.zeros((self.y.shape[0], padding_after,  self.y.shape[-1])) , ], axis=1 )

    # cette fonction doit renvoyer le nombre de batches réalisables par epoch avec les données du générateur
    def __len__(self): # should return the number of batches per epoch
        # l'argument allow_incomplete_batches permet de choisir si on autorise le renvoi d'un batch incomplet
        # par exemple si batch_size = 8 et qu'il y a 9 samples
        # si allow_incomplete_batches = False -> le générateur renvoie 1 batch de 8 par epoch
        # si allow_incomplete_batches = True -> le générateur renvoie 2 batches, le premier de 8, le deuxième de 1
        # globalement allow_incomplete_batches = True pour train,
        # et allow_incomplete_batches = False pour valid/test
        if self.allow_incomplete_batches:
            return int(np.ceil(self.x.shape[0]/self.batch_size))
        return self.x.shape[0]//self.batch_size

    # cette fonction est appelée par model.fit() à la fin de chaque epoch
    # l'objectif est de 'réinitialiser' le générateur afin qu'il re-mélange les échantillons
    # pour les présenter dans un ordre différent et avec une augmentation différente
    def on_epoch_end(self): # shuffle samples
        self.indexes = np.arange(self.x.shape[0])
        if self.shuffle:
            self.rng.shuffle(self.indexes)

    # cette fonction permet de renvoyer un batch (x et y) pour l'entraînement ou l'inférence
    def __getitem__(self, index):
        # get indexes for current batch
        b_start = index*self.batch_size
        b_end = np.minimum(((index+1)*self.batch_size), len(self.indexes))
        batch_indexes = self.indexes[b_start:b_end]
        
        # get data
        batch_x = self.x[batch_indexes,...].copy()
        batch_y = self.y[batch_indexes,...].copy()
        
        # normalisation
        batch_x = batch_x / batch_x.max(axis=1)[:,None,:]
        
        # augment id needed
        if self.augmenter is not None:
            batch_x, batch_y = self.augmenter.augment(batch_x, batch_y)
        
        # reshape in 3D
        # take reference curves, repeat them and stack them with GAMkl curves
        batch_x = np.concatenate([np.expand_dims(np.tile(batch_x[...,[0]],5),-1), np.expand_dims(batch_x[...,1:],-1)], -1)
        
        # ci-dessous : le même code que la ligne juste au-dessus, mais en version "étape par étape", plus compréhensible :
        if False:
            elp_curves = batch_x[...,[0]] # on extrait uniquement les courbes ELP, shape = (N,300,1)
            elp_curves = np.expand_dims(elp_curves, axis=-1) # on rajoute 1 dimension -> (N,300,1,1)
            elp_curves = np.repeat(elp_curves, 5, axis=-2) # on répète les ELP 5 fois (1 pour chaque G, A, M, K, L) -> (N,300,5,1)

            gamkl_curves = batch_x[...,1:] # on extrait les courbes G,A,M,K,L (N,300,5)
            gamkl_curves = np.expand_dims(gamkl_curves, axis=-1) # on rajoute 1 dimension -> (N,300,5,1)

            # on stacke GAMKL et ELP -> (N,300,5,2)
            batch_x = np.concatenate([elp_curves, gamkl_curves], axis = -1)
            
        # on fait la même chose pour y : c'est plus simple car il faut juste changer la dimension pour passer de (N,300,5) à (N,300,5,1)
        batch_y = np.expand_dims(batch_y, axis=-1)
        
        return batch_x, batch_y
```

## 2. Création du modèle et entraînement

### 2.1. Création du modèle

En premier pour cette partie, nous allons créer la structure du modèle. D'abord, on crée des fonctions qui vont permettre d'instancier le modèle, en fonction des différents hyperparamètres choisis en début de script.

Notre modèle sera un simple CNN basé sur la structure U-net.

```python
# la fonction pour ajouter un bloc de convolution à la structure U-net
def conv1d_block(input_tensor, n_filters, kernel_size_width=3, kernel_size_height=1, batchnorm=True):
    # first layer
    #x = Conv1D(filters=n_filters, kernel_size=kernel_size, kernel_initializer="he_normal", padding="same", data_format='channels_last')(input_tensor)
    x = tf.keras.layers.Conv2D(filters=n_filters, kernel_size=(kernel_size_width,kernel_size_height), kernel_initializer="he_normal", padding="same", data_format='channels_last')(input_tensor)
    if batchnorm:
        x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("relu")(x)
    # second layer
    #x = Conv1D(filters=n_filters, kernel_size=kernel_size, kernel_initializer="he_normal", padding="same", data_format='channels_last')(x)
    x = tf.keras.layers.Conv2D(filters=n_filters, kernel_size=(kernel_size_width,kernel_size_height), kernel_initializer="he_normal", padding="same", data_format='channels_last')(x)
    if batchnorm:
        x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("relu")(x)
    return x

def get_unet_custom(input_signal, blocks=4, wide_kernel_starts_at_block=-1, n_filters=16, kernel_size=3, dropout=0.1, batchnorm=True, n_classes=5, contract=1, return_model=True, return_sigmoid=True):
    # contracting path
    conv_blocks = []

    # encoder
    last_pb = input_signal
    for b in range(blocks):
        kernel_size_height = 1
        if (wide_kernel_starts_at_block>-1) and (b>=wide_kernel_starts_at_block-1):
            kernel_size_height=5
        dropout_mult=1.
        if b==0:
            dropout_mult=.5
        print("Adding conv layer in block {} with kernel size ({},{})".format(b+1,kernel_size,kernel_size_height))
        cb = conv1d_block(last_pb, n_filters=n_filters*np.power(2,b), kernel_size_width=kernel_size, kernel_size_height=kernel_size_height, batchnorm=batchnorm)
        conv_blocks.append(cb)
        if b<(blocks-1):
            pb = tf.keras.layers.MaxPooling2D((2,1)) (cb)
            pb = tf.keras.layers.Dropout(dropout*dropout_mult)(pb)
            last_pb = pb

    # decoder
    for b in range(blocks-1):
        ub = tf.keras.layers.Conv2DTranspose(n_filters*np.power(2,blocks-b-2), (kernel_size,1), strides=(2,1), padding='same') (conv_blocks[-1])
        ub = tf.keras.layers.Concatenate() ([ub, conv_blocks[blocks-b-2]])
        ub = tf.keras.layers.Dropout(dropout)(ub)
        cb = conv1d_block(ub, n_filters=n_filters*np.power(2,blocks-b-2), kernel_size_width=kernel_size, batchnorm=batchnorm)
        conv_blocks.append(cb)

    outputs = tf.keras.layers.Conv2D(n_classes, (1,contract), activation='sigmoid') (conv_blocks[-1])
    model = tf.keras.models.Model(inputs=[input_signal], outputs=[outputs])
    return model
```

Puis on peut instancier le modèle proprement dit en lui fournissant les hyperparamètres choisis :

```python
# on crée le placeholder pour l'input
input_signal = tf.keras.layers.Input((SPE_WIDTH, 5, 2), name='input_if')

# on construit le modèle depuis l'input
model = get_unet_custom(input_signal,
                        blocks=BLOCKS,
                        wide_kernel_starts_at_block=WIDE_KERNEL_STARTS_AT_BLOCK,
                        n_filters=FILTERS,
                        kernel_size=KERNEL_SIZE,
                        dropout=DROPOUT,
                        batchnorm=BATCHNORM,
                        n_classes=1,
                        contract=1)

# on crée le placeholder pour l'input
input_signal = tf.keras.layers.Input((SPE_WIDTH, 5, 2), name='input_if')

# on crée des callbacks pour monitorer l'entraînement :
# - model checkpoint -> permet de sauvegarder les poids du modèle quand la val loss s'améliore
# - reduce LR on plateau -> permet de diminuer le learning rate quand la val loss stagne (3 epochs sans amélioration)
# - early stopping -> permet d'arrêter l'entraînement quand la val loss stagne (10 epochs sans amélioration)
callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, min_delta=1e-3, verbose=VERBOSE, restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3, min_delta=1e-3, min_lr=MIN_LR, verbose=VERBOSE),
    tf.keras.callbacks.ModelCheckpoint(os.path.join(PATH_OUT,MODEL_NAME), monitor='val_loss', verbose=VERBOSE, save_best_only=True, save_weights_only=False)
]

# on crée notre propre métrique (IoU):
def curve_iou(y_true, y_pred, smooth = 1e-5):
    trh = tf.cast(tf.greater(y_true, .5), 'double')
    prd = tf.cast(tf.greater(y_pred, .5), 'double')
    i = tf.cast(tf.greater(trh+prd, 1), 'double')
    u = tf.cast(tf.greater(trh+prd, 0), 'double')
    i = tf.reduce_sum(i)
    u = tf.reduce_sum(u)
    return (smooth+i) / (smooth+u)
tf.keras.metrics.curve_iou=curve_iou

# on compile le modèle (i.e. on ajoute la loss, les métriques et on instancie l'algorithme d'optimisation)
model.compile(loss=tf.keras.losses.BinaryCrossentropy(), optimizer=tf.keras.optimizers.Adam(lr=BASE_LR), metrics=[curve_iou])
```

```python
print('Setting batch size to: '+str(BATCH_SIZE))
print('Setting maximal number of epochs to: '+str(N_EPOCHS))
print("")

# on lance l'entraînement proprement dit
results = model.fit(train_gen,
                    batch_size=BATCH_SIZE,
                    epochs=N_EPOCHS,
                    callbacks=callbacks,
                    verbose=VERBOSE,
                    validation_data=valid_gen,
                    validation_steps = valid_gen.__len__())

# à la fin de l'entraînement, on sauvegarde les logs (monitoring des losses/métriques)
logs = results.history
with open(os.path.join(PATH_OUT,LOG_NAME), 'wb') as file_pi:
    pickle.dump(logs, file_pi)
```

#### 3.3.2. Métriques de segmentation

On va calculer, pour chaque échantillon, deux métriques dont on affichera ensuite la moyenne :
1. L'IoU *réelle*, soit l'indice de Jaccard pour un seuil à 0.5 ; cette métrique peut être rapprochée de la précision  

2. L'IoU *tensorflow*, soit une approximation de l'IoU sans prendre en compte de seuil (i.e. uniquement du delta entre les valeurs prédites réellement positives et les valeurs prédites réellement négatives) ; cette métrique peut être rapprochée de l'AUC (peut être bonne même si le modèle ne renvoie que des valeurs supérieures ou inférieures à 0.5, tant que les valeurs pour les points positifs sont supérieures à celles pour les points négatifs)

```python
def metrics_curve_iou(y_true, y_pred, smooth = 1e-5):
    trh = tf.cast(tf.greater(y_true, .5), 'double')
    prd = tf.cast(tf.greater(y_pred, .5), 'double')
    i = tf.cast(tf.greater(trh+prd, 1), 'double')
    u = tf.cast(tf.greater(trh+prd, 0), 'double')
    i = tf.reduce_sum(i)
    u = tf.reduce_sum(u)
    return (smooth+i) / (smooth+u)

# we below compute accuracy (pr) and iou (both using our tensorflow method + a regular iou method without tensorflow)
# accuracy (pr) = percentage of correctly classified points
#     i.e. for each point among the N_SAMPLES*304*5 values in the dataset
#     we consider it correctly classified if round(predicted value) == ground truth value (if threshold==.5)
# iou : for each sample S and each dimension D (i.e., curve)
#     sum( prediction(S,D)>.5 AND ground truth(S,D)==1 ) / sum( prediction(S,D)>.5 OR ground truth(S,D)==1 )
threshold = .5
points=np.arange(1,SPE_WIDTH+1,1)
iou=np.zeros((ymap_test_proc_.shape[0],5))
iou_tf=np.zeros((ymap_test_proc_.shape[0],5))
for ix in trange(ymap_test_proc_.shape[0]):
    for dim in range(5):
        gt = ymap_test_proc[ix,:,dim,0] # get ground truth annotations for this sample & this curve
        pd_ = (ymap_test_proc_[ix,:,dim,0]>threshold)*1 # get predictions for this sample & this curve (transform into 0/1 according to selected threshold)
        u = np.sum(gt+pd_>0) # union : either predictions OR ground truth == 1
        i = np.sum(gt+pd_==2) # intersect : both predictions AND ground truth == 1
        if np.isfinite(u):
            iou[ix,dim] = i/u
        else: # u==0 -> division by zero
            iou[ix,dim] = np.nan
        iou_tf[ix,dim] = metrics_curve_iou(ymap_test_proc[ix,:,dim,0].astype('double'),ymap_test_proc_[ix,:,dim,0].astype('double'))

# print results        
print("")

# iou "réelle", i.e. points > ou < au seuil, vs. points à 0 ou 1
print("Mean IoU (real, vs. treshold={}):".format(threshold))
for k in range(iou.shape[1]):
    print("    > Fraction '{}': {:.2f} +- {:.2f}".format(['G','A','M','k','l'][k],np.nanmean(iou[:,k]),np.nanstd(iou[:,k])))

# l'iou calculée par méthodes tensorflow est plus intéressante car elle n'utilise pas de seuil
# mais tient compte uniquement de l'ordre des valeurs (i.e. similaire à une AUC)
print("Mean IoU (tensorflow, threshold-independant):")
for k in range(iou_tf.shape[1]):
    print("    > Fraction '{}': {:.2f} +- {:.2f}".format(['G','A','M','k','l'][k],np.nanmean(iou_tf[:,k]),np.nanstd(iou_tf[:,k])))
```

#### 3.3.3. Métriques de classification

On a convertir les maps de segmentation en classes, puis déterminer un tableau de contingence (comme pour le script de classification).

En premier, on écrit notre fonction qui convertit des maps en classe textuelle :

```python
# pour chaque échantillon, on va d'abord déterminer une classe "ground truth"
# en fonction des anomalies retrouvées
def classifySample(postprocessed_predicted_curves):
    """
    cette fonction prend en entrée une np.array de dimension 304x5 (les 5 courbes post processed prédites par le modèle)
    et renvoie une str: la classe de l'échantillon
    peut marcher à la fois pour les annotatinos réelles (ground truth) et les prédictions (si seuillées)
    """
    sample_class = "Unknown"
    # on détermine les deltas
    # par exemple pour une section de courbe:
    #  0  0  1  1  1  1  0  0
    # on va obtenir les nouvelles valeurs suivantes :
    #     0  1  0  0  0 -1  0
    # on a donc une nouvelle courbe (pour chaque dimension, i.e. G, A, M, k, l)
    # qui nous renseigne sur l'entrée/sortie d'un pic:
    # valeur==1 : on rentre dans un pic à cette position de la courbe
    # valeur==-1 : on sort d'un pic à cette position de la courbe
    # valeur==0 : on reste dans/hors d'un pic, pas de changement à cette position
    delta_curves = np.diff(postprocessed_predicted_curves, axis=0)
    # à partir de ces deltas, pour chaque dimension, on peut déterminer la position à laquelle on entre dans un pic
    mspikes_start_positions = [np.where(delta_curves[...,d]==1)[0] for d in range(delta_curves.shape[-1])]
    # on en déduit, pour chaque dimension, le nombre de pics retrouvés
    mspikes_numbers = [len(m) for m in mspikes_start_positions]
    # plusieurs cas de figures à ce stade:
    #    aucun pic sur la courbe : tous les 'mspikes_numbers' sont ==0
    #    1 pic sur la courbe :
    #        1 ou plusieurs 'mspikes_numbers' sont == 1 et toutes les valeurs 'mspikes_start_positions' sont
    #        égales à 0 ou la même valeur (i.e. les pics commencent au même endroit)
    #    plusieurs pics sur la courbe :
    #        plusieurs 'mspikes_numbers' sont == 1 et toutes les valeurs 'mspikes_start_positions' qui
    #        ne sont pas égales à zéro ont au moins 2 valeurs différentes (par ex: un pic commence à 200, et un autre commence à 240)
    # on va analyser ces différents cas de figure
    if all([m == 0 for m in mspikes_numbers]):
        # aucun pic: nombre de mspike==0 pour toutes les dimensions
        sample_class = "Normal"
    else: # au moins un pic détecté
        # on va faire la liste de toutes les "start positions" détectées, pour toutes les dimensions (G, A, M, k, l)
        start_positions_detected = [pos for m in mspikes_start_positions for pos in m]
        # on commence par voir s'il n'y aurait qu'une seule dimension (i.e. chaîne) affectée
        if len(start_positions_detected) == 1:
            # c'est le cas : juste un pic, donc probablement une chaîne légère libre ?
            # on vérifie qu'il n'y a pas de chaîne lourde, comme attendu
            if all(np.array(mspikes_numbers[:3])==0):
                # ok, c'est donc bien une chaîne légère libre
                light_concerned = np.where(np.array(mspikes_numbers[3:])==1)[0][0]
                sample_class = "Free_"+("K","L")[light_concerned]
            else:
                # il n'y a qu'une seule courbe affectée, et c'est une chaîne lourde -> ce n'est pas normal !
                sample_class = "Complex**"
        else:
            # sinon, on va déterminer si cette "position" est toujours la même sur les différentes chaînes
            all_mspikes_start_at_same_position = all([pos == start_positions_detected[0] for pos in start_positions_detected[1:]])
            if all_mspikes_start_at_same_position:
                # tous les pics commencent au même endroit, donc il n'y a qu'un seul pic monoclonal
                # il ne nous reste plus qu'à déterminer de quel type de chaînes il s'agit
                heavy_concerned = np.where(np.array(mspikes_numbers[:3])>0)[0]
                light_concerned = np.where(np.array(mspikes_numbers[3:])>0)[0]
                if (len(heavy_concerned) > 1) | (len(light_concerned) > 1):
                    # plusieurs chaînes lourdes ou légères pour le même pic ? probablement une erreur
                    sample_class = "Complex*"
                elif (len(heavy_concerned) == 0) & (len(light_concerned) == 1): # pas de chaine lourde mais une chaîne légère -> pic à chaîne légère libre
                    sample_class = "Free_"+("k","l")[light_concerned[0]]
                elif (len(heavy_concerned) == 1) & (len(light_concerned) == 1): # 1 chaîne lourde, 1 chaîne légère -> cas classique
                    sample_class = "Ig"+("G","A","M")[heavy_concerned[0]]+("_K","_L")[light_concerned[0]]
                else: # normalement on ne devrait pas arriver là (sauf si heavy==0 et light==0)
                    # mais on prévoit le coup au cas où
                    sample_class = "Error 1"
            else:
                # il y a visiblement plusieurs pics
                # on va déterminer s'il s'agit toujours de la même Ig
                # ou s'il s'agit au moins une fois d'une Ig différente
                # pour cela, on va simplement comparer les courbes G, A, M, k, l qui ne sont pas remplies de zéros entre elles
                # si elles sont identiques : c'est toujours la même Ig
                # sinon : le pic n'est pas toujours composé des mêmes chaînes
                ref_mspikes_loc = None
                all_affected_chains_are_identical=True
                for d in range(len(mspikes_start_positions)): # pour chaque courbe (i.e., dimension, G, A, M, k, l)
                    if len(mspikes_start_positions[d])==0:
                        continue # pas de pic sur cette courbe -> on continue
                    else: # >= 1 pic a été trouvé
                        if ref_mspikes_loc is None:
                            # on avait pas encore trouvé de pic
                            ref_mspikes_loc = mspikes_start_positions[d]
                        else: # on avait déjà trouvé une chaîne avec 1/des pics
                            if len(ref_mspikes_loc) == len(mspikes_start_positions[d]):
                                if any(ref_mspikes_loc != mspikes_start_positions[d]):
                                    # les pics ne sont pas à la même position sur cette chaîne!
                                    all_affected_chains_are_identical = False
                                    break
                            else:
                                all_affected_chains_are_identical = False
                                break
                if all_affected_chains_are_identical:
                    # un seul "type" de pics : on peut la classer dans une catégorie "basique"
                    heavy_concerned = np.where(np.array(mspikes_numbers[:3])>0)[0]
                    light_concerned = np.where(np.array(mspikes_numbers[3:])>0)[0]
                    if (len(heavy_concerned) > 0) & (len(light_concerned) > 0): # pics à Ig entière
                        sample_class = "Ig"+("G","A","M")[heavy_concerned[0]]+("_K","_L")[light_concerned[0]]
                    elif (len(heavy_concerned) == 0) & (len(light_concerned) > 0): # pas de chaine lourde mais >=1 chaîne légère -> pics à chaîne légère libre
                        sample_class = "Free_"+("K","L")[light_concerned[0]]
                    else: # normalement on ne devrait pas arriver là (heavy>0 et light==0 ?)
                        # mais on prévoit le coup au cas où
                        sample_class = "Error 2"
                else:
                    # plusieurs types de pics : on va la classer en "complexe"
                    sample_class = "Complex"
    return sample_class
```

Puis on peut appeler cette fonction pour chaque échantillon afin d'obtenir la "classe prédite" pour chaque échantillon.

Pour les classes réelles (*ground truth*), on n'utilisera pas cette fonction mais directement les annotations textuelles originales chargées dans la première partie de ce script (cela permet d'être complètement sûr qu'il n'y a aucune erreur dans la conversion pour les labels réelles et que l'on estime correctement les performances du modèle).

```python
sample_pred_classes = []

# the threshold used for the predictions // should be left at 50% by default
threshold=.5

# pour chaque courbe on calcule :
for i in tqdm(range(ymap_test_proc.shape[0])):
    # la classe prédite ; que l'on stocke dans une list []
    sample_pred_classes.append(classifySample((ymap_test_proc_[i,...,0]>threshold)*1))
    
# il ne nous reste plus qu'à convertir cette liste en numpy array (post processing plus aisé après)
sample_pred_classes = np.array(sample_pred_classes)

# pour les annotations réelles, on copie simplement notre liste déjà chargée depuis le fichier .csv
sample_gt_classes = y_test.copy()
```

Enfin, on va pouvoir calculer et afficher notre tableau de contingence :

```python
# We can write a simple cross tab with pandas
contingency_table = pd.crosstab(sample_gt_classes, sample_pred_classes)
contingency_table.index.name = "Ground truth:"
contingency_table.columns.name = "Predicted:"

# on calcule la précision par classe "ground truth" :
pr_by_groundtruth_class = np.array([contingency_table.loc[c,c] for c in contingency_table.index]) / np.sum(np.array(contingency_table), axis=1)
# et on ajoute ça en colonne au tableau de contingence :
contingency_table.loc[:,"Accuracy (by ground truth class)"] = ["{:.1f} %".format(100*v) for v in pr_by_groundtruth_class]

pr = np.sum(np.array(sample_pred_classes)==np.array(sample_gt_classes))/len(sample_gt_classes) # on calcule la précision globale

# on calcule les vrais & faux positifs et négatifs (tp, tn, fp, fn) en fonction de "normal" ou pas "normal"
tp = np.sum( (np.array(sample_pred_classes)!="Normal") & (np.array(sample_gt_classes)!="Normal") )
fp = np.sum( (np.array(sample_pred_classes)!="Normal") & (np.array(sample_gt_classes)=="Normal") )
tn = np.sum( (np.array(sample_pred_classes)=="Normal") & (np.array(sample_gt_classes)=="Normal") )
fn = np.sum( (np.array(sample_pred_classes)=="Normal") & (np.array(sample_gt_classes)!="Normal") )
# on peut alors calculer sensibilité & spécificité pour la question "Normal vs. non Normal"
se = tp/(tp+fn)
sp = tn/(tn+fp)

display(contingency_table)

print("Précision globale : {:.1f} %".format(100*pr))
print("")
print("Normal vs. anormal :")
print("    Sensibilité : {:.1f} %".format(100*se))
print("    Spécificité : {:.1f} %".format(100*sp))
```
